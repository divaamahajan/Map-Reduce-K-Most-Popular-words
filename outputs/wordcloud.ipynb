```python
import pandas as pd

# Read the text file
# df = pd.read_csv('top100Output.txt', delimiter='\t')
df = pd.read_csv('top100Output.txt', delimiter='\t', encoding='utf-16', header=None)

df[1] = pd.to_numeric(df[1], errors='coerce').astype('Int64')

# Display the DataFrame
df.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Assuming you have frequencies as a DataFrame or Series
frequencies = df.set_index(0)[1].to_dict()

wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(frequencies)

# Plot the word cloud with legend
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
